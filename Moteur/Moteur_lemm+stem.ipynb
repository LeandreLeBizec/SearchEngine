```python
import pandas as pd
import numpy as np
```

```python
numreq=30
numdoc=1460
```

## Lecture des vecteurs

```python
invreq = open("lemstemreq.txt","r").read().replace(";]",'').replace("]",'').split("\n")
invreq.pop(len(invreq)-1)
for i in range(len(invreq)) :
    invreq[i]=invreq[i].split(" :[")
for i in range(len(invreq)) :
    invreq[i][1]=invreq[i][1].split(';')
    for j in range(len(invreq[i][1])) :
        invreq[i][1][j]=invreq[i][1][j].replace(' ','').split(':')

invreq
```

```python
vecreq =np.zeros((numreq,len(invreq)))

for i in range(len(invreq)) :
    for val in invreq[i][1] :
        if val[0] :
            vecreq[int(val[0])-1][i]=float(val[1])
        
vecreq
```

```python
len(vecreq)
```

```python
invdoc = open("lemstemdoc.txt","r").read().replace(";]",'').replace("]",'').split("\n")
invdoc.pop(len(invdoc)-1)
for i in range(len(invdoc)) :
    invdoc[i]=invdoc[i].split(" :[")
for i in range(len(invdoc)) :
    invdoc[i][1]=invdoc[i][1].split(';')
    for j in range(len(invdoc[i][1])) :
        invdoc[i][1][j]=invdoc[i][1][j].replace(' ','').split(':')
invdoc
```

```python
vecdoc =np.zeros((numdoc,len(invdoc)))

for i in range(len(invdoc)) :
    for val in invdoc[i][1] :
        if val[0] :
            vecdoc[int(val[0])-1][i]=float(val[1])
        
vecdoc
```

```python
len(vecdoc)
```

## Calcul du cosinus

```python
from scipy import spatial
cos = []
for i in range(len(vecreq)) :
    cos.append([])
    for j in range(len(vecdoc)) :
        cos[i].append(spatial.distance.cosine(vecreq[i], vecdoc[j]))

cos
```

## Liste des documents pour chaque requte

```python
res={}
for i in range(numreq) :
    res[str(i+1)]=[]
    for j in range(numdoc) :
        if (cos[i][j] < 0.849) & (cos[i][j] != 0) :
            score= 1-cos[i][j]
            #print(score)
            res[str(i+1)].append(str(j+1)+"\t"+str(score))
            #0.849, f max
res
```

## Exportation des rsultats

```python
file = open("RES_LemmStem.REL","w")

for cle, val in res.items() :
    dico=[]
    for v in val :
        
        v_split= v.split("\t")
        dico.append(v_split)
    array=np.array(dico)  
    print(array)
    
    if len(array) != 0:
        array=array[np.lexsort((array[:, 1], ))]
        array=array[::-1]
    for doc in array:
        file.write(str(cle)+"\t"+str(doc[0])+"\t"+str(doc[1])+"\n")
file.close()
```
